```python
import cv2
import torch
from ultralytics import YOLO

# Load YOLOv8 model (pre-trained on COCO dataset)
model = YOLO("yolov8n.pt")  # You can train a custom model for better accuracy

# Load the video
video_path = "/content/cow counting vedio.mp4"  # Change to your video path
cap = cv2.VideoCapture(video_path)

# Get video properties
fps = int(cap.get(cv2.CAP_PROP_FPS))
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fourcc = cv2.VideoWriter_fourcc(*"mp4v")
output_path = "/content/Cow_Counting.mp4"
out = cv2.VideoWriter(output_path, fourcc, fps, (width, height))

# Define the class ID for cows (Check YOLO's dataset for accuracy)
COW_CLASS_ID = 19  # This ID may vary depending on the dataset used

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # Run YOLO object detection
    results = model(frame)

    cow_count = 0  # Initialize cow count

    for result in results:
        for box in result.boxes:
            class_id = int(box.cls[0].item())  # Get detected class ID
            x1, y1, x2, y2 = map(int, box.xyxy[0])  # Bounding box coordinates
            label = result.names[class_id]  # Class label

            # Check if the detected object is a cow
            if class_id == COW_CLASS_ID or "cow" in label.lower():
                cow_count += 1
                color = (0, 255, 0)  # Green box for cows
            else:
                color = (0, 0, 255)  # Red box for other objects

            # Draw bounding box and label
            cv2.rectangle(frame, (x1, y1), (x2, y2), color, 2)
            cv2.putText(frame, f"{label}", (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, color, 2)

    # Display cow count on the frame
    cv2.putText(frame, f"Total Cows: {cow_count}", (50, 50), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 255), 2)

    # Write frame to output video
    out.write(frame)

cap.release()
out.release()

print("Processing complete! Download the processed video here:")
print(output_path)
```
