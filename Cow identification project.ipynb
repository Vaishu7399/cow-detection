```python
import cv2
import torch
from ultralytics import YOLO

# Load YOLOv8 model (pre-trained on COCO dataset)
model = YOLO("yolov8n.pt")

# Load the video
video_path = "/content/cow identification vedio .2.mp4"
cap = cv2.VideoCapture(video_path)

# Get video properties
fps = int(cap.get(cv2.CAP_PROP_FPS))
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fourcc = cv2.VideoWriter_fourcc(*"mp4v")
output_path = "/content/processed_video.mp4"
out = cv2.VideoWriter(output_path, fourcc, fps, (width, height))

# Class ID for cow (Check YOLO class list for accuracy)
COW_CLASS_ID = 19

def identify_animals():
    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break

        # Perform object detection
        results = model(frame)

        cow_count = 0
        other_animals = 0

        for result in results:
            for box in result.boxes:
                class_id = int(box.cls[0].item())
                x1, y1, x2, y2 = map(int, box.xyxy[0])
                label = result.names[class_id]

                # Identify cow or other animals
                if class_id == COW_CLASS_ID or "cow" in label.lower():
                    cow_count += 1
                    color = (0, 255, 0)  # Green for cows
                    text = "Cow"
                else:
                    other_animals += 1
                    color = (0, 0, 255)  # Red for other animals
                    text = f"{label}"

                # Draw bounding box and label
                cv2.rectangle(frame, (x1, y1), (x2, y2), color, 2)
                cv2.putText(frame, text, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.7, color, 2)

        # Display counts
        cv2.putText(frame, f"Cows: {cow_count}", (50, 50), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
        cv2.putText(frame, f"Other: {other_animals}", (50, 90), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2)

        # Write frame to output
        out.write(frame)

    cap.release()
    out.release()
    print("Processing complete! Download the processed video here:", output_path)

identify_animals()
```
